# Dealing with Null Values

```python
dataframe.isnull().sum()
```

```python
# Data Cleaning
# dropping all missing values rows

dataframe.dropna(inplace=True)

dataframe.isnull().sum()
```

# Dealing with Duplicate values

```python
dataframe.duplicated().sum()
```

```python
# deleting all duplicated rows
# data[data.duplicated()]

dataframe.drop_duplicates(inplace=True)
dataframe.duplicated().sum()
```

```python
dataframe.info()
```

```python
dataframe.shape
```

# Type casting 

```python

# changing data types 

dataframe.dtypes

dataframe['id'] = dataframe['id'].astype(object)
dataframe.dtypes

dataframe['host_id'] = dataframe['host_id'].astype(object)
dataframe.dtypes
```

# Data Analysis

```python
# indetifying outliers 
df = dataframe[dataframe['price']< 1600]
sns.boxplot(data=df, x='price')
```

```python
# Price distribution
sns.histplot(data= df,x='price')
```

```python
#Price distribuion

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='price', bins=100)
plt.title('Price Distribuition')
plt.ylabel("Frequency")
plt.show()
```

```python
df.dtypes
```

```python
#Price distribuion
plt.figure(figsize=(6, 3))
sns.histplot(data=df, x='availability_365')
plt.title('availability_365 Distribuition')
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

# Feature Engineering

```python
df['price per bed']= df['price']/df['beds']
df.head()
```

```python
# average price per bed
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```
